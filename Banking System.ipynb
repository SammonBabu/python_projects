Write a python program to replicate a Banking system. The following features are mandatory:
1. Account login
2. Amount Depositing
3. Amount Withdrawal
Other than the above features you can add any other also.

```python
import random

class AccountDetails:
    def __init__(self, name, user_name, password, account_number):
        self.name = name
        self.user_name = user_name
        self.password = password
        self.account_number = account_number
        self.balance = 0

class Account:
    def __init__(self):
        self.accounts = {}

    def sign_up(self, name, user_name, password):
        account_number = random.randint(1000000000, 9999999999)
        if account_number in self.accounts:
            return self.sign_up(name, user_name, password)
        self.accounts[account_number] = AccountDetails(name, user_name, password, account_number)
        return account_number

    def login(self, user_name, password):
        for account in self.accounts.values():
            if user_name == account.user_name and password == account.password:
                return True
        return False

    def logout(self):
        return True

    def deposit(self, account_number, amount):
        self.accounts[account_number].balance += amount
        print(f"Deposited {amount} successfully")
        return self.accounts[account_number].balance

    def withdraw(self, account_number, amount):
        if amount > self.accounts[account_number].balance:
            return 'Insufficient funds'
        self.accounts[account_number].balance -= amount
        print(f"Withdrawn {amount} successfully")
        return self.accounts[account_number].balance

    def check_balance(self, account_number):
        return self.accounts[account_number].balance
    
    def transfer(self, account_number, amount, receiver_account_number):
        if amount > self.accounts[account_number].balance:
            return 'Insufficient funds'
        self.accounts[account_number].balance -= amount
        self.accounts[receiver_account_number].balance += amount
        print(f'Transferred {amount} successfully to {receiver_account_number}')
        return self.accounts[account_number].balance
    
    def main(self):
        print('*' * 30)
        print('Welcome to the banking system')
        print('*' * 30)
        while True:
            print('Please select an option')
            print('1. Sign up')
            print('2. Login')
            print('3. Exit')
            print('*' * 30)
            choice = int(input('Enter your choice: '))

            if choice == 1:
                name = input('Enter your name: ')
                user_name = input('Enter your user name: ')
                password = input('Enter your password: ')
                account_number = self.sign_up(name, user_name, password)
                print(f'Account created successfully, your account number is {account_number}')

            elif choice == 2:
                user_name = input('Enter your user name: ')
                password = input('Enter your password: ')
                if self.login(user_name, password):
                    print('Login successful')
                    account_number = None
                    for acc_num, account in self.accounts.items():
                        if account.user_name == user_name:
                            account_number = acc_num
                            break
                    print(f'Welcome {self.accounts[account_number].name}')
                    print(f'Your account number is {account_number}')
                    print(f'Your current balance is {self.accounts[account_number].balance}')
                    while True:
                        print('*' * 30)
                        print('1. Deposit')
                        print('2. Withdraw')
                        print('3. Check balance')
                        print('4. Transfer')
                        print('5. Logout')
                        choice = int(input('Enter your choice: '))
                        if choice == 1:
                            amount = int(input('Enter the amount to deposit: '))
                            print(f'Your current balance is {self.deposit(account_number, amount)}')

                        elif choice == 2:
                            amount = int(input('Enter the amount to withdraw: '))
                            print(f'Your current balance is {self.withdraw(account_number, amount)}')
                        elif choice == 3:
                            print(f'Your current balance is {self.check_balance(account_number)}')
                        elif choice == 4:
                            receiver_account_number = int(input('Enter the receiver account number: '))
                            amount = int(input('Enter the amount to transfer: '))
                            print(f'Your current balance is {self.transfer(account_number, amount, receiver_account_number)}')
                        elif choice == 5:
                            self.logout()
                            print('Logout successful')
                            break
                        else:
                            print('Invalid choice')

                else:
                    print('Invalid credentials')

            elif choice == 3:
                print('Thank you for using the banking system')
                break

            else:
                print('Invalid choice')

account = Account()
account.main()
```
